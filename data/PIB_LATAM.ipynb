```python
df_PIB_paises_trend = pd.concat([df_PIB_paises_2022,df_PIB_paises_2020, df_PIB_paises_2018,df_PIB_paises_2016,df_PIB_paises_2014], ignore_index=True)
df_PIB_paises_trend["año"] = df_PIB_paises_trend.index.to_series().apply(
    lambda x: 2024 if x <= 5
    else 2022 if 6 <= x <= 10
    else 2020 if 11 <= x <= 15
    else 2018 if 16 <= x <= 20
    else 2016 if 21 <= x <= 25
    else 2014)
df_PIB_paises_trend.set_index('año', inplace=True)
df_PIB_paises_trend.sort_values(by=["País", "año"], ascending=False, inplace=True)
display(df_PIB_paises_trend)
```

```python
from scipy.stats import linregress

df_PIB_paises_trend = df_PIB_paises_trend.reset_index()
df_PIB_paises_trend["año"] = pd.to_numeric(df_PIB_paises_trend["año"], errors="coerce")
df_PIB_paises_trend["value"] = pd.to_numeric(df_PIB_paises_trend["value"], errors="coerce")

# Cálculo de la pendiente Peru (m)
slope_per, _, r_value_per, _, _ = linregress(df_PIB_paises_trend.loc[1:5, "año"], df_PIB_paises_trend.loc[1:5, "value"])
print(f"Pendiente Peru: {slope_per:.2f}, R² = {r_value_per**2:.3f}")

# Cálculo de la pendiente Mexico (m)
slope_mex,_, r_value_mex, _, _ = linregress(df_PIB_paises_trend.loc[6:10, "año"], df_PIB_paises_trend.loc[6:10, "value"])
print(f"Pendiente Mexico: {slope_mex:.2f}, R² = {r_value_mex**2:.3f}")

# Cálculo de la pendiente Costa Rica (m)
slope_cr,_, r_value_cr, _, _ = linregress(df_PIB_paises_trend.loc[11:15, "año"], df_PIB_paises_trend.loc[11:15, "value"])
print(f"Pendiente Costa Rica: {slope_cr:.2f}, R² = {r_value_cr**2:.3f}")

# Cálculo de la pendiente Colombia (m)
slope_col,_, r_value_col, _, _ = linregress(df_PIB_paises_trend.loc[16:20, "año"], df_PIB_paises_trend.loc[16:20, "value"])
print(f"Pendiente Colombia: {slope_col:.2f}, R² = {r_value_col**2:.3f}")

# Cálculo de la pendiente Chile (m)
slope_chi,_, r_value_chi, _, _ = linregress(df_PIB_paises_trend.loc[21:25, "año"], df_PIB_paises_trend.loc[21:25, "value"])
print(f"Pendiente Chile: {slope_chi:.2f}, R² = {r_value_chi**2:.3f}")


# Cálculo de la pendiente Bol (m)
slope_bol,_, r_value_bol, _, _ = linregress(df_PIB_paises_trend.loc[26:30, "año"], df_PIB_paises_trend.loc[26:30, "value"])
print(f"Pendiente Bolivia: {slope_bol:.2f}, R² = {r_value_bol**2:.3f}")
```
